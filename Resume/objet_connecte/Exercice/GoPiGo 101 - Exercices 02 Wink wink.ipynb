### GoPiGo 101 - Série d'exercices 2
##### Manipulation des **phares**.
Les phares sont les 2 lumières rouges sont montées sur la plaquette rouge `Dexter GoPiGo3`. Ces deux diodes électroluminescentes sont situées sous la plaquette du robot à l'avant à gauche et à droite.

---

Les fonctions suivantes permettent la manipulation des phares :
 - `EasyGoPiGo3.blinker_on` ou `EasyGoPiGo3.led_on` : allume un phare
 - `EasyGoPiGo3.blinker_off` ou `EasyGoPiGo3.led_off` : éteint un phare

### Démonstration

```python
import easygopigo3 as gpg
import time

robot = gpg.EasyGoPiGo3()
time_to_wait = 1. # en secondes

# les méthodes suivantes sont des substituts équivalents :
#  - blinker_on <---> led_on
#  - blinker_off <---> led_off
# les paramètres suivants sont des substituts équivalents :
#  - 0 <---> 'right'        # entier ou chaine de caractères
#  - 1 <---> 'left'         # entier ou chaine de caractères

robot.blinker_on(0) # 0 = droite
robot.blinker_on(1) # 1 = gauche
time.sleep(time_to_wait)

robot.blinker_off('right') # int ou str
robot.blinker_off('left') # int ou str
time.sleep(time_to_wait)

robot.led_on(0) # 0 = droite
robot.led_on(1) # 1 = gauche
time.sleep(time_to_wait)

robot.led_off('right') # int ou str
robot.led_off('left') # int ou str
# on termine

del time_to_wait
del robot
del time
del gpg
```

### Exercice 2.2.
Développez une fonction qui permet de réaliser le même objectif que celui de l'exerice `2.1.`. Votre fonction doit avoir cette signature :

`simple_blink(for_n_seconds : int = 1) -> None`

`for_n_seconds` doit être obligatoirement un entier limité entre 1 et 60 inclusivement.

Testez votre fonction pour 5 secondes.

```python
# Solution
def simple_blink(for_n_seconds : int = 1) -> None :
    m = int(for_n_seconds * 2)
    if(m >= 1 and m <= 60):
        for i in range(m):
            robot.blinker_on(0)
            robot.blinker_on(1)
            time.sleep(0.25)
            robot.blinker_off(0)
            robot.blinker_off(1)
            time.sleep(0.25)
simple_blink(5)        
            
def simple_blink(for_n_seconds : int = 1) -> None:
    # validation de l'entier
    if not isinstance(for_n_seconds, int):
        raise ValueError('for_n_seconds must be an integer')
    
    # validation de l'interval
    # 3 approches différentes pour faire cette validation - les 3 approches sont équivalentes et donnnent le même résultat
    # approche 1
    if for_n_seconds < 1:
        for_n_seconds = 1
    if for_n_seconds > 60:
        for_n_seconds = 60
    # approche 2
    for_n_seconds = 1 if for_n_seconds < 1 else (60 if for_n_seconds > 60 else for_n_seconds) # les parenthèses sont inutiles
    # approche 3
    for_n_seconds = max(1, min(for_n_seconds, 60))

    # boucle de clignotement
    time_to_wait_by_cycle = 1.
    time_to_wait_half_cycle = time_to_wait_by_cycle / 2.
    for _ in range(for_n_seconds):
        robot.led_on(0)
        robot.led_on(1)
        time.sleep(time_to_wait_half_cycle)
        robot.led_off(0)
        robot.led_off(1)
        time.sleep(time_to_wait_half_cycle)
        
simple_blink(5)
```

### Exercice 2.3.
Développez une fonction qui permet de réaliser un clignotement un peu plus sophistiqué. Votre fonction doit avoir cette signature :

`cool_blink(n_times : int = 1, cycle_duration : float = 1., percent_on : float = 0.5) -> None`

où :
 - `n_times` indique le nombre de fois à clignoter
 - `cycle_duration` : indique la durée d'un cycle en secondes : cycle = ouvert->fermé
 - `percent_on` : indique le pourcentage de la durée ouverte vs fermée (en % entre 0.0 et 1.0)

avec ces contraintes :
 - `n_times >= 1`
 - `cycle_duration >= 0.1`
 - `0.0 <= percent_on <= 1.0`

Par exemple, l'appel suivant donne 3 clignotements d'une durée de 1.5 secondes où la lumière est ouverte à 75% du temps et fermée à 25% du temps. 

`cool_blink(3, 1.5, 0.75)`

Faites un test avec ces 3 appels de fontion :
 - `cool_blink(3, 1.5, 0.75)`
 - `cool_blink(5, 1.0, 0.5)`
 - `cool_blink(10, 0.35, 0.1)`

```python
# Solution
def cool_blink(n_times : int = 1, cycle_duration : float = 1., percent_on : float = 0.5) -> None:
    if n_times < 1 :
        n_times = 1
    if cycle_duration < 0.1:
        cycle_duration
    if percent_on < 0:
        percent_on = 0
    if percent_on > 1:
        percent_on = 1
    
    on_duration = cycle_duration * percent_on
    off_duration = 1 - percent_on
    for i in range(n_times):
        robot.blinker_on(0)
        robot.blinker_on(1)
        time.sleep(on_duration)
        robot.blinker_off(0)
        robot.blinker_off(1)
        time.sleep(off_duration)

    
cool_blink(3, 1.5, 0.75)
cool_blink(5, 1.0, 0.5)
cool_blink(10, 0.35, 0.1)

def cool_blink(n_times : int = 1, cycle_duration : float = 1., percent_on : float = 0.5) -> None:
    if not isinstance(n_times, int):
        raise ValueError('n_times must be an integer')
    if not isinstance(cycle_duration, float):
        raise ValueError('cycle_duration must be a real')
    if not isinstance(percent_on, float):
        raise ValueError('percent_on must be a real')
        
    n_times = 1 if n_times < 1 else n_times
    cycle_duration = 0.1 if cycle_duration < 0.1 else cycle_duration
    percent_on = max(0., min(percent_on, 1.))
    percent_off = 1.0 - percent_on
    
    duration_on = cycle_duration * percent_on
    duration_off = cycle_duration * percent_off
    
    for _ in range(n_times):
        robot.led_on(0)
        robot.led_on(1)
        time.sleep(duration_on)
        robot.led_off(0)
        robot.led_off(1)
        time.sleep(duration_off)
        
cool_blink(3, 1.5, 0.75)
cool_blink(5, 1.0, 0.5)
cool_blink(10, 0.35, 0.1)
```

### Exercice 2.5. §

Considérant la variable `time_to_wait` égale à `2.3725`, faites cette séquence avec le robot :
 1. avancez pendant `time_to_wait` secondes 
 1. arrêter pendant `time_to_wait / 2.` secondes
 1. reculez pendant `time_to_wait` secondes
 1. arrêter pendant `time_to_wait / 2.` secondes
 1. tournez à droite pendant `time_to_wait` secondes
 1. tournez à gauche pendant `time_to_wait` secondes
 1. arrêter le robot et terminer immédiatement le programme
    
Toutefois, vous devez faire clignoter les phares selon cette stratégie :
 - lorsque le robot est à l'arrêt, les phares sont éteints
 - lorsque le robot avance, les 2 phares clignotent à un cycle de 0.75 seconde et ouvert à 25% du temps
 - lorsque le robot recule, les 2 phares clignotent à un cycle de 0.25 seconde et ouvert à 75% du temps
 - lorsque le robot tourne, le phare du côté tournant clignote à un cycle de 0.5 seconde alors que le phare opposé est éteint

```python
# Solution
time_to_wait = 2.3725

# Fonction pour faire clignoter les phares selon les spécifications données
def clignoter_phare(cycle, cycle_duration):
    robot.set_eye_color((255, 0, 0))  # Définit la couleur des phares à rouge
    robot.led_on(0)  # Allume les phares
    robot.led_on(1) 
    time.sleep(cycle * cycle_duration)  # Temps d'allumage
    robot.led_off('right')  # Éteint les phares
    robot.led_off('left')  # Éteint les phares
    time.sleep(cycle * (1 - cycle_duration))  # Temps d'extinction


robot.forward()  # Avancer
clignoter_phare(0.75, 0.25)  # Clignotement pendant l'avance
time.sleep(time_to_wait)  # Attendre
robot.stop()  # Arrêter

clignoter_phare(0.0, 0.0)  # Éteindre les phares

robot.backward()  # Reculer
clignoter_phare(0.25, 0.75)  # Clignotement pendant le recul
time.sleep(time_to_wait)  # Attendre
robot.stop()  # Arrêter

clignoter_phare(0.0, 0.0)  # Éteindre les phares

robot.right()  # Tourner à droite
clignoter_phare(0.5, 1.0)  # Clignotement pendant la rotation à droite
time.sleep(time_to_wait)  # Attendre
robot.left()  # Tourner à gauche
clignoter_phare(0.5, 1.0)  # Clignotement pendant la rotation à gauche
time.sleep(time_to_wait)  # Attendre
robot.stop()
```

### Exercice 2.6. * §
La question 2.3. est relativement simple car il y a un nombre entier de cycle - 1 cycle, 2 cycles, 10 cycles ... Le défi suivant consiste à faire une fonction qui fait clignoter pendant un certain temps peu importe où le cycle est rendu lors de l'arrêt.

Développez une fonction qui permet de réaliser un clignotement selon cette signature :

`blink(total_duration : float, cycle_duration : float = 1., percent_on : float = 0.5, begin_on : bool = True, end_off : bool = True) -> None`

où :
 - `total_duration` indique la durée totale du clignotement
 - `cycle_duration` : indique la durée d'un cycle ouvert-fermé en secondes
 - `percent_on` : indique le pourcentage de la durée ouverte vs fermée (en % entre 0.0 et 1.0)
 - `begin_on` : indique si les phares sont ouverts ou fermés au début
 - `end_off` : indique si les phares sont ouverts ou fermés lorsque la fonction termine

avec ces contraintes :
 - `total_duration >= 0.1`
 - `cycle_duration >= 0.1`
 - `0.0 <= percent_on <= 1.0`

Cet exercice est plus difficile à ce moment-ci des exercices. Nous y reviendrons dans le projet final. L'important est d'au moins analyser cet énoncé et de comprendre les enjeux en place.

```python
# Solution
import time

def blink(total_duration: float, cycle_duration: float = 1.0, percent_on: float = 0.5, begin_on: bool = True, end_off: bool = True) -> None:
    # Vérification des contraintes
    if total_duration < 0.1 or cycle_duration < 0.1 or percent_on < 0.0 or percent_on > 1.0:
        raise ValueError("Les paramètres ne respectent pas les contraintes")

    # Calcul du nombre de cycles
    num_cycles = total_duration / cycle_duration

    # Détermination de l'état initial des phares
    if begin_on:
        # Si les phares doivent commencer allumés
        print("Allumer les phares")
    else:
        # Si les phares doivent commencer éteints
        print("Éteindre les phares")

    # Début du clignotement
    start_time = time.time()

    for _ in range(int(num_cycles)):
        # Allumer les phares pour le pourcentage spécifié
        print("Phares allumés")
        time.sleep(cycle_duration * percent_on)

        # Éteindre les phares pour le reste du cycle
        print("Phares éteints")
        time.sleep(cycle_duration * (1 - percent_on))

    # Calcul du temps restant
    remaining_time = total_duration - (num_cycles * cycle_duration)

    # Si la durée totale n'est pas un multiple de la durée du cycle
    if remaining_time > 0:
        # Allumer les phares pour le pourcentage spécifié pendant le temps restant
        print("Phares allumés (temps restant)")
        time.sleep(remaining_time * percent_on)

        # Éteindre les phares pour le reste du temps
        print("Phares éteints (temps restant)")
        time.sleep(remaining_time * (1 - percent_on))

    # Fin du clignotement
    end_time = time.time()

    # Détermination de l'état final des phares
    if end_off:
        # Si les phares doivent se terminer éteints
        print("Éteindre les phares (fin)")
    else:
        # Si les phares doivent se terminer allumés
        print("Allumer les phares (fin)")

    # Affichage de la durée totale écoulée
    print("Durée totale écoulée:", end_time - start_time, "secondes")

# Test de la fonction blink avec des paramètres arbitraires
blink(total_duration=5, cycle_duration=1, percent_on=0.7, begin_on=True, end_off=True)
```
